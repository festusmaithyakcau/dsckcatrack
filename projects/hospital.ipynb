Replacing some column data 

```python
df['sex'] = df['sex'].replace([1],'Male')
df.sex=df.sex.replace([0],"Female")
df
```

Abstracting columns from the dataframe

```python
trtbps = list(df.trtbps)
trtbps
```

```python
age = list(df.age)
age
```

```python
chol = list(df.chol)
chol
```

Visualizing the data using different tools

```python
plt.scatter(age, trtbps,label="Age vs trtbps",color="green")
plt.xlabel("Age")
plt.ylabel("trtbps")
plt.legend(loc="best")
plt.show()
```

```python
plt.scatter(age, chol, label='Age against Cholestrol' ,color="red")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.legend(loc="best")
plt.grid()
plt.show()
```

3D SCATTER PLOT

```python
sns.set_style("darkgrid")

mean = df.age.mean()
number = 50

x1 = np.random.normal(mean, 1, size=number)
y1 = np.random.normal(mean, 1, size=number)
z1 = np.random.normal(mean, 1, size=number)

plt.figure(figsize=(6, 5))
axes = plt.axes(projection='3d')
print(type(axes))
axes.scatter3D(x1, y1, z1, s=100)

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_zlabel('z')
plt.show()
```

Comparison of mean values 

```python
meantrtbps= df.trtbps.mean()
meanchol=df.chol.mean()
thalachhmean = df.thalachh.mean()
agemean=df.age.mean()


plt.close("all")
sizes=[meantrtbps,meanchol,thalachhmean,agemean]
labels=["trtbps","cholesterol","thalachh","Age"]
castout=[0,0.02,0,.2]
plt.figure(figsize=(5,5))
plt.pie(sizes, labels=labels,explode=castout, autopct='%.2f %% ')
plt.show()
```
