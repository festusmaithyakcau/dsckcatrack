```python
import seaborn as sns
import matplotlib.pyplot as plt
```

CREATING A SCATTER PLOT

```python
height = [62, 64, 69, 75, 66,
68, 65, 71, 76, 73]
weight = [120, 136, 148, 175, 137,
165, 154, 172, 200, 187]
sns.scatterplot(x=height, y=weight)
plt.show()
```

CREATING A COUNT PLOT 

```python
gender = ["Female","Female","Female","Female","Male","Male","Male","Male","Male","Male"]
sns.countplot(x=gender)
plt.show()
```

Working with DataFrames

```python
import pandas as pd
df = pd.read_csv("masculinity.csv")
df.head()
```

```python
df = pd.read_csv("masculinity.csv")
sns.countplot(x="q0002",data=df)
plt.show()
```

Using hue property

```python
sns.scatterplot(x="age3",y="weight",data=df,hue="q0001")
plt.show()
```

Setting hue order

```python
sns.scatterplot(x="age3",y="weight",data=df,hue="q0001",hue_order=["Very masculine","Not at all masculine"])
plt.show()
```

Specifying hue colors

```python
hue_colors = {"No answer": "black"
,
"Not very masculine": "red"}

sns.scatterplot(x="age3",y="weight",data=df,hue="q0001",hue_order=["Not very masculine","No answer"],palette=hue_colors)
plt.show()
```

Using HTML hex color codes with hue

```python
hue_colors = {"No answer": "#808080"
,
"Not very masculine": "#00FF00"}

sns.scatterplot(x="age3",y="weight",data=df,hue="q0001",hue_order=["Not very masculine","No answer"],palette=hue_colors)
plt.show()
```

Using hue with count plots

```python
sns.countplot(x="weight",data=df,hue="educ3")
plt.show()
```
