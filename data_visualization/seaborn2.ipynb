```python
#impoting essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading dataset
df = seaborn.load_dataset("tips")
df
```

```python
 # first five rows of dataset
df.head()
```

```python
 # last five rows of dataset
df.tail()
```

Swarm Plots


```python
# use swarmplot to visualize tip observations and amounts
# by day of the week
sns.swarmplot(x="day", y="tip", data=df)
sns.set_style("whitegrid")
plt.show()
```

```python
# visualize tip observations
sns.swarmplot(x=df["tip"])
sns.set_style("darkgrid")
plt.show()
```

```python
# color points by category
# create customized palette
gender_palette = ["#A833FF", "#FFAF33"]
sns.swarmplot(x="day", y="tip", hue="sex", palette=gender_palette, data=df)
plt.show()
```

```python
 # control plot order on x-axis
sns.swarmplot(x="smoker", y="total_bill", data=df,  hue="sex",palette="husl", order=["Yes", "No"])
plt.show()
```

Violin Plots

```python
# plot tips
sns.violinplot(x = df["tip"], color="gold")
plt.show()
```

```python
# draw plot based on variable
sns.violinplot(x ="sex",y ="total_bill",data=df)
plt.show()
```

```python
# Split drawings to compare with hue/legend variables
sns.violinplot(x = "time",y ="tip",data=df, hue ="sex",palette ="dark",split = True)
plt.legend()
plt.show()
```

Facet Grids


```python
# draw facet grid based on tip variable
fg = sns.FacetGrid(df,col = "time",row = "sex")

fg = fg.map(plt.hist,"tip", color ="tomato")
```

```python
# we can also change the type of plot and the colors around the points
fg = sns.FacetGrid(df, col="time", row="sex")
fg = fg.map(plt.scatter, "total_bill", "tip", color="floralwhite", edgecolor="hotpink")
```

```python
# plot by category
x = sns.FacetGrid(df, col="time", hue="sex")
x = x.map(plt.scatter,"total_bill","tip")
x =x.add_legend()
```
