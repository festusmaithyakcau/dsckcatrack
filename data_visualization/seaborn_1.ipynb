```python
#impoting essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading dataset
df = sns.load_dataset("tips")
df
```

```python
 # first five rows of dataset
df.head()
```

```python
 # last five rows of dataset
df.tail()
```

Swarm Plots


```python
# use swarmplot to visualize tip observations and amounts
# by day of the week
sns.swarmplot(x="day", y="tip", data=df)
sns.set_style("whitegrid")
plt.show()
```

```python
# visualize tip observations
sns.swarmplot(x=df["tip"])
sns.set_style("darkgrid")
plt.show()
```

```python
# color points by category
# create customized palette
gender_palette = ["#A833FF", "#FFAF33"]
sns.swarmplot(x="day", y="tip", hue="sex", palette=gender_palette, data=df)
plt.show()
```

```python
 # control plot order on x-axis
sns.swarmplot(x="smoker", y="total_bill", data=df,  hue="sex",palette="husl", order=["Yes", "No"])
plt.show()
```

Violin Plots

```python
# plot tips
sns.violinplot(x = df["tip"], color="gold")
plt.show()
```

```python
# draw plot based on variable
sns.violinplot(x ="sex",y ="total_bill",data=df)
plt.show()
```

```python
# Split drawings to compare with hue/legend variables
sns.violinplot(x = "time",y ="tip",data=df, hue ="sex",palette ="dark",split = True)
plt.legend()
plt.show()
```

Facet Grids


```python
# draw facet grid based on tip variable
fg = sns.FacetGrid(df,col = "time",row = "sex")

fg = fg.map(plt.hist,"tip", color ="tomato")
```

```python
# we can also change the type of plot and the colors around the points
fg = sns.FacetGrid(df, col="time", row="sex")
fg = fg.map(plt.scatter, "total_bill", "tip", color="floralwhite", edgecolor="hotpink")
```

```python
# plot by category
x = sns.FacetGrid(df, col="time", hue="sex")
x = x.map(plt.scatter,"total_bill","tip")
x =x.add_legend()
```

3D SCATTER PLOTS

```python
mean = 3
number =50

x1 = np.random.normal(mean, 1, size=number)
y1 = np.random.normal(mean, 1, size=number)
z1 = np.random.normal(mean, 1, size=number)

# Plotting the 3D scatter
plt.figure(figsize=(6,5))
axes = plt.axes(projection='3d')
```

```python
print(type(axes))
axes.scatter3D(x1, y1, z1)

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_zlabel('z')
plt.show()
```

```python
sns.set_style("darkgrid")

mean = 3
number =50

x1 = np.random.normal(mean, 1, size=number)
y1 = np.random.normal(mean, 1, size=number)
z1 = np.random.normal(mean, 1, size=number)

plt.figure(figsize=(6,5))
axes = plt.axes(projection='3d')
print(type(axes))
axes.scatter3D(x1, y1, z1)

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_zlabel('z')
plt.show()
```

```python
# Increasing the size of the scatter plots
sns.set_style("darkgrid")

mean = 3
number =50

x1 = np.random.normal(mean, 1, size=number)
y1 = np.random.normal(mean, 1, size=number)
z1 = np.random.normal(mean, 1, size=number)

plt.figure(figsize=(6,5))
axes = plt.axes(projection='3d')
print(type(axes))
axes.scatter3D(x1, y1, z1,s=100)

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_zlabel('z')
plt.show()
```

```python
# A spiral Plot
sns.set_style("whitegrid")

OMEGA = 2

Z1 = np.linspace(0, 10, 100)
X1 = np.cos(OMEGA*Z1)
Y1= np.sin(OMEGA*Z1)

plt.figure(figsize=(6,5))
axes = plt.axes(projection='3d')
axes.plot3D(X1,Y1,Z1)
# keeps padding between figure elements
plt.tight_layout()
plt.show()
```

```python
# a 3D plot of a map

def finalfunc(x, y):
    return 50 - (x**2 + y**2)


sns.set_style('whitegrid')
N = 50

X_VAL = np.linspace(-5, 5, N)
Y_VAL = np.linspace(-5, 5, N)

X1, Y1 = np.meshgrid(X_VAL, Y_VAL)

Z1 =finalfunc(X1, Y1)

axes = plt.axes(projection='3d')
axes.plot_surface(X1, Y1, Z1)
plt.show()
```
